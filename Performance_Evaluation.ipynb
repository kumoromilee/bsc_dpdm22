```python
from sklearn import datasets
iris_data = datasets.load_iris()
```

```python
X = iris_data['data']
```

```python
y = iris_data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

```

```python
import numpy as np
```

```python
# Define the number of folds  กำหนดจำนวนของ folds
n_folds = 10

# Split the data into features and target  แยกข้อมูลเป็น features and target
X = shuffled_df.drop('label', axis=1)
y = shuffled_df['label']

# Define a cross-validation object with the specified number of folds กำหนด  cross-validation object ด้วยspecified number of folds
kfold = KFold(n_splits=n_folds)

# Loop over the folds วนลูป #ภายในลูปจะใช้วิธีการแยก KFold เพื่อรับค่าของ training and testing data จากนั้นจะใช้ค่านี้เพื่อแบ่งข้อมูลออกเป็น test and train 
for fold, (train_idx, test_idx) in enumerate(kfold.split(X, y)):
    # Get the training and testing data for the current fold
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]
    
    # Create a decision tree classifier object สร้างต้นไม้ตัดสินใจ
    dtc = DecisionTreeClassifier()
    
    
    # Fit the model on the training data
    dtc.fit(X_train, y_train)
    
    # Evaluate the model on the testing data ประเมิน model ด้วยtesting data
    scores = cross_val_score(dtc, X_test, y_test, cv=kfold)
    
    # Print the cross-validation score and the data for the current fold
    print(f"Fold: {fold+1}")
    print("Mean accuracy: %0.2f (+/- %0.2f)" % (np.mean(scores), np.std(scores) * 2))
    print(shuffled_df[shuffled_df['fold'] == fold+1])
    print()
```
